```python
import numpy as np
```

```python
import random as r
sample = [] #list for the 1% low 
l = 180 #180 seconds or 3 minutes of benchmark
av = []
for i in range(181):
    L = r.randint(55,60)
    A = r.randint(115,120)
    sample +=[L]
    av += [A] 
    if i is 179:
        sample+=[20]
        av += [120]
        break

```

```python
print(sample)
print(av)
```

```python
r.shuffle(sample)
r.shuffle(av)
avgFPS = sum(av)/len(av)
avg = sum(sample)/len(sample)
print(sample)#1% lows
print()
print(av)#avg Framerate
print()
print(avgFPS,'FPS')
print(avg,'1% low FPS')
```

```python
#now see what the average is its 57.42.... and we have a 20 FPS spike which will be very noticeable as a stutter
import matplotlib.pyplot as plt 
frametimes = [(1/x)*1000 for x in sample]
fig = plt.figure(figsize = (10,10))
plt.xlabel('time (seconds)',fontsize=12)
plt.ylabel('frametime (ms)',fontsize=12)
plt.xlim()
plt.ylim()
plt.plot(frametimes)
plt.show()
print(avg,'1% low FPS')
print(avgFPS,'average FPS')
```

```python
#here how this looks on a bar graph
top = [('avg FPS',avgFPS),('1% low',avg),(' ',0)]
labels, ys = zip(*top)
xs = np.arange(len(labels)) 
width = 0.5
fig = plt.figure(figsize = (10,10))
plt.bar(xs, ys, width, align='center')
plt.xticks(xs, labels) 
plt.yticks(ys)
plt.show()
```
